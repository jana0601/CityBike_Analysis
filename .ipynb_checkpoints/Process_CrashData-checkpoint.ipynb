```python
import pandas as pd
accident_path = r"C:\Users\yanhu\Documents\Python_Notebook\AXA_task\accident_data\Motor_Vehicle_Collisions_-_Crashes_20251009.csv"

# Load full CSV, combine date and time into one datetime column while reading
accident_df = pd.read_csv(
    accident_path,
    parse_dates=[['CRASH DATE', 'CRASH TIME']]
)

accident_df.head(20)
# Rename the combined column
accident_df.rename(columns={'CRASH DATE_CRASH TIME': 'CRASH_DATETIME'}, inplace=True)

# Sort by datetime
accident_df.sort_values('CRASH_DATETIME', inplace=True)

# Reset index after sorting
accident_df.reset_index(drop=True, inplace=True)

# Save sorted data to new CSV
sorted_csv_path = 'sorted_accident_data.csv'  # change path as needed
accident_df.to_csv(sorted_csv_path, index=False)

print(f"Sorted data saved to: {sorted_csv_path}")
```

```python
# Filter for accidents involving cyclists
sorted_bike_crashes = accident_df[
    (accident_df['NUMBER OF CYCLIST INJURED'] > 0) |
    (accident_df['NUMBER OF CYCLIST KILLED'] > 0)
]
# Save sorted data to new CSV
sorted_csv_path_bike = ' r"C:\Users\yanhu\Documents\Python_Notebook\AXA_task\accident_data\sorted_bike_crashes.csv'  # change path as needed
sorted_bike_crashes.to_csv(sorted_csv_path_bike, index=False)
```
