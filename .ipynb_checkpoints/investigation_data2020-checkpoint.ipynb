```python
import pandas as pd
#trip_path = r"C:\Users\zhao\Documents\workspace\task\merged_tripdata2\2025_merged.csv"
trip_df = pd.read_csv(trip_path) #, nrows=80000
trip_df
```

```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
from pathlib import Path

# === Parse datetime ===
trip_df['starttime'] = pd.to_datetime(trip_df['starttime'], errors='coerce')

# === Filter valid start coordinates ===
trip_points = trip_df[
    trip_df['start station latitude'].notna() & trip_df['start station longitude'].notna()
].copy()

# === Optional: group by month if needed ===
trip_points['year_month'] = trip_points['starttime'].dt.to_period('M').astype(str)

# === Map setup ===
map_center = [
    trip_points['start station latitude'].mean(),
    trip_points['start station longitude'].mean()
]
m = folium.Map(location=map_center, zoom_start=12)

# === Marker cluster for trip start points ===
trip_cluster = MarkerCluster(name="Trip Start Locations").add_to(m)

# === Add markers ===
for _, row in trip_points.iterrows():
    popup_text = (
        f"<b>Bike ID:</b> {row['bikeid']}<br>"
        f"<b>Start Station:</b> {row['start station name']}<br>"
        f"<b>Start Time:</b> {row['starttime']}"
    )
    folium.Marker(
        location=[row['start station latitude'], row['start station longitude']],
        popup=popup_text,
        icon=folium.Icon(color='blue', icon='bicycle', prefix='fa')
    ).add_to(trip_cluster)

# === Add layer control ===
folium.LayerControl(collapsed=False).add_to(m)

# === Save the map ===
output_file = Path("results/trip_start_locations_map2020.html")
output_file.parent.mkdir(parents=True, exist_ok=True)
m.save(str(output_file))
print(f"âœ… Trip map saved to: {output_file}")
```
